```python
def morgan_fpts(data):
    Morgan_fpts = []
    for mol in data:
        fpts = AllChem.GetMorganFingerprintAsBitVect(mol,2,1024)
        mfpts = np.array(fpts)
        Morgan_fpts.append(mfpts)
    return np.array(Morgan_fpts)
```

```python
suppl = Chem.SDMolSupplier('concatFinal.sdf',removeHs=False, sanitize=False)
morgan_fpts(suppl)
```
